#### Machine Learning Clustering Analysis

```python
X = df2
y = df2['Complain'] #assigning feature matrix and target
```

```python
# Checking the informations of the Independent variables

X.info()
```

```python
# assigning the column features to the cariable cols
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])  # Create a dataframe from the feature matrix
```

```python
X.head() # view the first five rows of the dataframe
```

```python
# Kmeans clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0) 

kmeans.fit(X)
```

#### K-Means model parameters study 

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

 #### Checking the quality of the weak classification by the model

```python
labels = kmeans.labels_

# check how many of the samples were correctly labeled
correct_labels = sum(y == labels)

print("Output: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
#### Checking model accuracy score

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

#### Using the elbow method to find the optimal number of clusters 

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2,random_state=0)

kmeans.fit(X)

labels = kmeans.labels_

# check how many of the samples were correctly labeled

correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)

kmeans.fit(X)

# check how many of the samples were correctly labeled
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
```

#### Using Scipy's hierarchy.linkage() to form clusters and plot them with hierarchy.dendrogram():

```python
clusters = hierarchy.linkage(df2, method="ward")

plt.figure(figsize=(8, 6))
dendrogram = hierarchy.dendrogram(clusters)
# Plotting a horizontal line based on the first biggest distance between clusters 
plt.axhline(150, color='red', linestyle='--'); 
# Plotting a horizontal line based on the second biggest distance between clusters 
plt.axhline(100, color='crimson'); 
```

```python
clustering_model = AgglomerativeClustering(n_clusters=2, linkage="ward")
clustering_model.fit(df2)
labels = clustering_model.labels_
```

#### Clustering the number of Complaints that were made from Web purchases

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
sns.scatterplot(ax=axes[0], data=df, x='NumWebPurchases', y='Complain').set_title('Without clustering')
sns.scatterplot(ax=axes[1], data=df, x='NumWebPurchases', y='Complain', hue=clustering_model.labels_).set_title('With clustering');
```

#### From Our AgglomerativeClustering we can see that based on our clusters 0 and 1 more customers did not complain from purchases made from the web

#### The  AgglomerativeClustering is suitable for our data because we can see how it adapts well with our model instances and forms a flexible and informative clusters
